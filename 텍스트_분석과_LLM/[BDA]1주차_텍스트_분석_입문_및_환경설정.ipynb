```python
# 변수는 데이터를 저장하는 이름표입니다
이름 = "김철수"
나이 = 25
키 = 175.5
학생여부 = True

print(이름)   # 김철수
print(나이)   # 25
```

# 리스트 : 여러 데이터를 순서대로 담기

```python
# 텍스트 분석에서 가장 많이 쓰이는 자료형입니다
리뷰들 = ["정말 맛있어요", "별로예요", "최고입니다", "다시는 안 사요"]

# 첫 번째 리뷰 꺼내기 (인덱스는 0부터 시작!)
print(리뷰들[0])   # 정말 맛있어요
print(리뷰들[1])   # 별로예요

# 전체 길이 확인
print(len(리뷰들))  # 4

# 리뷰 추가하기
리뷰들.append("또 주문할게요")
print(len(리뷰들))  # 5
```

# 딕셔너리 : 이름표를 붙여서 데이터 저장하기

```python
# 키(key): 값(value) 형태로 저장
리뷰 = {
    "텍스트": "정말 맛있어요",
    "별점": 5,
    "작성자": "홍길동"
}

print(리뷰["텍스트"])  # 정말 맛있어요
print(리뷰["별점"])    # 5
```

# 반복문 : 같은 작업을 자동으로 반복

```python
리뷰들 = ["정말 맛있어요", "별로예요", "최고입니다"]

# 리뷰 하나씩 출력하기
for 리뷰 in 리뷰들:
    print("리뷰:", 리뷰)

# 결과:
# 리뷰: 정말 맛있어요
# 리뷰: 별로예요
# 리뷰: 최고입니다
```

# 함수 : 자주 쓰는 코드를 묶어두기

```python
# 함수 정의
def 인사하기(이름):
    print(f"안녕하세요, {이름}님!")

# 함수 사용
인사하기("김철수")   # 안녕하세요, 김철수님!
인사하기("이영희")   # 안녕하세요, 이영희님!
```

# 종합 실습 : 리뷰 분석 흉내내기

```python
# 리뷰 데이터
리뷰_목록 = [
    "너무 맛있어요 최고",
    "별로예요 실망입니다",
    "정말 맛있고 좋아요",
    "다시는 안 사겠습니다",
    "강력 추천합니다"
]

# 긍정 키워드 리스트
긍정_키워드 = ["맛있", "최고", "좋아", "추천"]

# 각 리뷰를 긍정/부정으로 분류
for 리뷰 in 리뷰_목록:
    분류 = "부정 😤"
    for 키워드 in 긍정_키워드:
        if 키워드 in 리뷰:
            분류 = "긍정 😊"
            break
    print(f"[{분류}] {리뷰}")

# 결과:
# [긍정 😊] 너무 맛있어요 최고
# [부정 😤] 별로예요 실망입니다
# [긍정 😊] 정말 맛있고 좋아요
# [부정 😤] 다시는 안 사겠습니다
# [긍정 😊] 강력 추천합니다
```

```python
from konlpy.tag import Okt

okt = Okt()
테스트_문장 = "나는 텍스트 분석을 배우고 있습니다"

# 형태소 분석 맛보기
print(okt.morphs(테스트_문장))

# 결과 (예시):
# ['나', '는', '텍스트', '분석', '을', '배우고', '있습니다']
```
