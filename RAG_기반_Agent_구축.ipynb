### (1) Vector DB мӨҖл№„

```python
from langchain_huggingface import HuggingFaceEmbeddings # OpenAI лҢҖмӢ  мӮ¬мҡ©
from langchain.vectorstores import Chroma
from langchain.document_loaders import CSVLoader

# CSV нҢҢмқј лЎңл“ң
csv_path = "sample.csv"
csv_loader = CSVLoader(file_path= path + csv_path)
documents_csv = csv_loader.load()
# Hugging Face мһ„лІ л”© лӘЁлҚё
embedding = HuggingFaceEmbeddings(model_name="jhgan/ko-sroberta-multitask")

vector_db = Chroma.from_documents(documents_csv, embedding,
                                    persist_directory= path + "chroma_db")
retriever = vector_db.as_retriever(search_kwargs={"k": 3})
print("ChromaDBм—җ м ҖмһҘ мҷ„лЈҢ!")
```

### (2) Tool м •мқҳ

```python
@tool
def vectordb_search(query: str) -> str:
    """лӮҙмһҘ л¬ём„ңлЎңл¶Җн„° кҙҖл Ё м •ліҙлҘј кІҖмғүн•©лӢҲлӢӨ."""  # docstring кјӯ нҸ¬н•Ёлҗҳм–ҙм•ј н•Ё!
    # мң мӮ¬лҸ„ кё°мӨҖмңјлЎң мғҒмң„ kк°ң л¬ём„ңлҘј лІЎн„°мҠӨнҶ м–ҙм—җм„ң кІҖмғү
    docs = retriever.get_relevant_documents(query)

    # кІҖмғүлҗң л¬ём„ң лӮҙмҡ©л“Өмқ„ мӨ„л°”кҝҲмңјлЎң кө¬л¶„н•ҳм—¬ н•ҳлӮҳмқҳ л¬ёмһҗм—ҙлЎң л°ҳнҷҳ
    return "\n\n".join(d.page_content for d in docs)

# tool лҰ¬мҠӨнҠё
tools = [vectordb_search]

# нҲҙ мӢӨн–ү л…ёл“ң (LangGraph м ңкіө)
tool_node = ToolNode(tools)

```

```python
# кө¬кёҖ кІҖмғүлҸ„кө¬ кө¬м„ұ : SerpAPI
tools_web = load_tools(["serpapi"])  # langchain лӮҙмһҘ лҸ„кө¬
web_search_tool = tools_web[0]  # serpapi tool к°қмІҙ
```

```python
# нҲҙ лҰ¬мҠӨнҠё
tools = [vectordb_search, web_search_tool]
```

### (3) LLM+Tool л°”мқёл”©

```python
tool_node = ToolNode(tools)
hf_ep = HuggingFaceEndpoint(repo_id="openai/gpt-oss-20b", task="conversational")
llm = ChatHuggingFace(llm=hf_ep).bind_tools(tools)
```

### (4) л…ёл“ң мӨҖл№„ & л¶„кё° н•ЁмҲҳ

```python
class State(TypedDict):
    messages: Annotated[list, add_messages]
```

```python
# GPT нҳём¶ң л…ёл“ң
def call_model(state: State):
    messages = state["messages"]
    response = llm.invoke(messages)

    # SQLite м ҖмһҘ мӢң л°ңмғқн•ҳлҠ” м§Ғл ¬нҷ” м—җлҹ¬ л°©м§Җ (HuggingFace нҠ№нҷ”)
    if hasattr(response, "additional_kwargs"):
        response.additional_kwargs = {}

    return {"messages": [response]}

# нҲҙ мӮ¬мҡ© м—¬л¶Җ нҢҗлӢЁ
def should_continue(state: State) -> Literal["tools", END]:
    last = state["messages"][-1]            # лҢҖнҷ”мқҳ к°ҖмһҘ л§Ҳм§Җл§ү к°қмІҙ(к°ҖмһҘ мөңк·ј лҢҖнҷ”)
    # if getattr(last, "tool_calls", None):   # lastм—җм„ң tool_calls лӮҙмҡ©мқҙ нҸ¬н•Ёлҗҳл©ҙ, 'tool_calls', м•„лӢҲл©ҙ None
    #     return "tools"
    if hasattr(last, "tool_calls") and last.tool_calls:
        return "tools"
    return END
```

### (5) к·ёлһҳн”„ кө¬м„ұ

```python
builder = StateGraph(State)
builder.add_node("call_model", call_model)
builder.add_node("tools", tool_node)

# connect nodes
builder.add_edge(START, "call_model")
builder.add_conditional_edges("call_model", should_continue)
builder.add_edge("tools", "call_model")

graph = builder.compile()
```
